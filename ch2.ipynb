```python
import nltk
from nltk.corpus import gutenberg
import matplotlib.pyplot as plt
%matplotlib inline

nltk.download('gutenberg')
```

```python
caesar = gutenberg.open('shakespeare-caesar.txt')
caesar = caesar.readlines()
caesar[:5]
```

```python
sentences = list(filter(None, [sent.strip("\n") for sent in caesar]))
sentences
```

```python
plt.hist([len(s) for s in sentences])
```

```python
# most common words
 
from collections import Counter
import re

words = [word for sentence in sentences for word in sentence.split()]
alphanumeric_words = list(filter(None, [re.sub('[^A-Za-z0-9]', '', w) for w in words]))

# remove stop words 
stopwords = nltk.corpus.stopwords.words('english')
no_stopwords = [w for w in alphanumeric_words if w not in stopwords]

counter = Counter(no_stopwords)
counter.most_common(10)
```
